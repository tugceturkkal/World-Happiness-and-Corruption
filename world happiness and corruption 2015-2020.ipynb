### Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix as cm
import seaborn as sns
sns.set()
```

### Load the data

```python
raw_data = pd.read_csv('WorldHappiness_Corruption_2015_2020.csv')
raw_data.head()
```

```python
data = raw_data.copy()


data['happiness_category'] = data['happiness_score']

# Countries with a happiness score greater than 5 were categorized as yes(happy), 
# and countries with less happiness score were categorized as no(unhappy).

for i in data['happiness_score']:
    if i < 5:
        data['happiness_category'] = data['happiness_category'].replace([i], 'No')
    else:
        data['happiness_category'] = data['happiness_category'].replace([i], 'Yes')
    
data

# Split data by years

data_2015 = data.iloc[0:132 , :]
data_2020 = data.iloc[660:, :]
data_2020
```

### Multiple Linear Regression

```python
y = data_2020['happiness_score']
x1 = data_2020[['health','freedom','social_support','cpi_score']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

### Logistic Regression

```python
data['y'] = data['happiness_category'].map({'Yes' : 1 , 'No' : 0})
data_2020 = data.iloc[660:, :]
data_2020
```

```python
y_logit = data['y']
x1_logit = data[['health','freedom','social_support','cpi_score']]
```

```python
x_logit = sm.add_constant(x1_logit)
regression_logit = sm.Logit(y_logit,x_logit)
results_logit = regression_logit.fit()
```

```python
results_logit.summary2()
```

```python
def confusion_matrix(data,actual_values,model):
    pred_values = model.predict(data)
    bins = np.array([0,0.5,1])
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
    accuracy = (cm[0,0] + cm[1,1]) / cm.sum()
    return cm, accuracy
```

```python
confusion_matrix(x_logit,y_logit,results_logit)
```

### Test the model 

```python
x_train , x_test , y_train , y_test = train_test_split(x_logit,y_logit,test_size = 0.2,random_state = 42)
logistic_reg = LogisticRegression(random_state=42)
logistic_reg.fit(x_train, y_train)
```

```python
predictions = logistic_reg.predict(x_test)
score = round(accuracy_score(y_test, predictions), 3)
cm1 = cm(y_test, predictions)
sns.heatmap(cm1, annot=True, fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score), size = 15)
plt.show()
```
