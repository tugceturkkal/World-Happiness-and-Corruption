### Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load the data

```python
raw_data = pd.read_csv('WorldHappiness_Corruption_2015_2020.csv')
raw_data.head()
```

```python
raw_data.describe()
```

```python
data = raw_data.copy()

# Countries with a happiness score greater than 5 were categorized as yes(happy), 
# and countries with less happiness score were categorized as no(unhappy).

for i in data['happiness_score']:
    if i < 5:
        data['happiness_score'] = data['happiness_score'].replace([i], 'No')
    else:
        data['happiness_score'] = data['happiness_score'].replace([i], 'Yes')
```
