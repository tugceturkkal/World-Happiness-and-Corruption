### Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

### Load the data

```python
raw_data = pd.read_csv('WorldHappiness_Corruption_2015_2020.csv')
raw_data.head()
```

```python
data = raw_data.copy()


data['happiness_category'] = data['happiness_score']

# Countries with a happiness score greater than 5 were categorized as yes(happy), 
# and countries with less happiness score were categorized as no(unhappy).

for i in data['happiness_score']:
    if i < 5:
        data['happiness_category'] = data['happiness_category'].replace([i], 'No')
    else:
        data['happiness_category'] = data['happiness_category'].replace([i], 'Yes')
    
data

# Split data by years

data_2015 = data.iloc[0:132 , :]
data_2020 = data.iloc[660:, :]
data_2020
```

### Multiple Linear Regression

```python
y = data_2020['happiness_score']
x1 = data_2020[['health','freedom','social_support','cpi_score']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

### Logistic Regression

```python
data['y'] = data['happiness_category'].map({'Yes' : 1 , 'No' : 0})
data_2020 = data.iloc[660:, :]
data_2020
```

```python
y_logit = data['y']
x1_logit = data[['health','freedom','social_support','cpi_score']]
```

```python
x_logit = sm.add_constant(x1_logit)
regression_logit = sm.Logit(y_logit,x_logit)
results_logit = regression_logit.fit()
results_logit.summary()
```
