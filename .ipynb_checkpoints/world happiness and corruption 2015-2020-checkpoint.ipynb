### Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

### Load the data

```python
raw_data = pd.read_csv('WorldHappiness_Corruption_2015_2020.csv')
raw_data.head()
```

```python
data = raw_data.copy()

# Split data by years

data_2015 = data.iloc[0:132 , :]
data_2020 = data.iloc[660:, :]
```

### Multiple Linear Regression

```python
y = data_2020['happiness_score']
x1 = data_2020[['health','freedom','government_trust','dystopia_residual','social_support','cpi_score']]
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```
